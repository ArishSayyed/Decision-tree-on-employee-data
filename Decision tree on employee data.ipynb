```python
import pandas as pd
import numpy as np
```

# Loading & Preparing the Data

```python
df = pd.read_csv("employee.csv")
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
#Finding the unique elements in various variable
print(df.EmployeeCount.unique())
print(df.EmployeeNumber.unique())
print(df.Over18.unique())
print(df.StandardHours.unique())
```

```python
#Removing the less significant variables
df = df.drop(["EmployeeCount","EmployeeNumber","Over18","StandardHours"],axis=1)
```

```python
df.columns
```

```python
#Checking NA values
df.isna().sum()
```

```python
#Replacing "ATTRITION" by Integer constant(Yes=1 & No=0)

df.loc[df["Attrition"]=="No", "Attrition"]=0

df.loc[df["Attrition"]=="Yes", "Attrition"]=1
```

```python
#Extracting NUMERIC variables
df_num = df[df.select_dtypes(np.number).columns.tolist()]
df_num.head()
```

```python
#Creating DUMMY variables
df_dummy = pd.get_dummies(df[df.select_dtypes(np.object).columns.tolist()])
df_dummy.head()
```

```python
#Concatinating both NUMERIC and CATEGORICAL variables
df_concat = pd.concat([df_num,df_dummy], axis=1)
df_concat.head()
```

```python
#Defining DEPENDENT and INDEPENDENT variables
x = df_concat.drop(["Attrition"], axis=1)
x.head()
```

```python
y = df_concat["Attrition"]
y.head()
```

```python
#Splitting Dataset into Training and Testing data
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3, random_state = 20)
```

# Training the model and making the Predictions

```python
from sklearn import tree
```

```python
dt = tree.DecisionTreeClassifier(criterion= "entropy", max_depth=3, random_state=20)

#Training the model
dt.fit(x_train, y_train)
```

```python
#Predicting the response for the Test Data
y_pred = dt.predict(x_test)
```

# Evaluating performance of the model.


```python
from sklearn import metrics
```

```python
#ACCURACY
print("Accuracy: ",metrics.accuracy_score(y_test,y_pred))
```

```python
#PRECISION
print("Precision: ",metrics.precision_score(y_test,y_pred))
```

```python
#RECALL
print("Recall: ",metrics.recall_score(y_test,y_pred))
```
